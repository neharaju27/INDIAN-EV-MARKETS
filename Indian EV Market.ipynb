### Converting date column to datetime and droping invalid dates

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
file_path = "ev_cat_01-24.csv" 
df1 = pd.read_csv(file_path)
```

```python
df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')
df1 = df1.dropna(subset=['Date'])
df1.set_index('Date', inplace=True) 
```

### Number of Registrations Done Over Time

```python
df1['Total_Registrations'] = df1.sum(axis=1)
print("\nTotal EV Registrations Done Over Time:\n", df1['Total_Registrations'].tail())
```

### Distribution Based On Catergory

```python
category_distribution = df1.sum().sort_values(ascending=False)
print("\nCategory-Wise EV Distribution:\n", category_distribution)
```

### Growth Rate of EV Adoption

```python
df1['Growth_Rate'] = df1['Total_Registrations'].pct_change() * 100
print("\nGrowth Rate of EV Adoption:\n", df1['Growth_Rate'].tail())
```

### Market Shares According To Vehicle Types

```python
market_share = (category_distribution / category_distribution.sum()) * 100
print("\nMarket Share of Vehicle Types:\n", market_share)
```

### Sudden Spikes or Drops

```python
threshold = df1['Total_Registrations'].std() * 2
anomalies = df1[np.abs(df1['Total_Registrations'] - df1['Total_Registrations'].mean()) > threshold]
print("\nAnomalies in Registrations:\n", anomalies)
```

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
df1 = df1[~((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis=1)]
plt.figure(figsize=(12, 5))
sns.boxplot(data=df1)
plt.xticks(rotation=55)
plt.title("Boxplot of Numerical Features")
plt.show()
```

```python
missing_df2 = df2.isnull().sum()
missing_df2
```

### Handling Inconsistent Data

```python
# Categorical columns
print(f"Cat Unique:")
print(df2['Cat'].unique())
print(f"Maker Unique:")
print(df2['Maker'].unique())
```

### Remove extra space using strip method

```python
df2['Maker'] = df2['Maker'].str.strip().str.title()
```

### Checking for Duplicates

```python
duplicates = df2.duplicated()
duplicates
```

### Reshaping the Data

```python
df_melted = df2.melt(id_vars=['Cat', 'Maker'], var_name='Year', value_name='Value')
df_melted['Year'] = df_melted['Year'].astype(int)
df_melted = df_melted.sort_values(by=['Maker', 'Year'])
df_melted
```

# Key Performance indicators of the given datasets.

### Total Units Sold Per year

```python
total_sales_per_year = df2.iloc[:, 2:].sum()
print("\nTotal Sales Per Year:\n", total_sales_per_year,)
```

### Growth Rate Every Year

```python
yearly_growth_rate = total_sales_per_year.pct_change().fillna(0) * 100
print("\nYear-over-Year Growth Rate (%):\n", yearly_growth_rate)
```

### Top Manufacturers

top_manufacturers = df2.groupby("Maker").sum().iloc[:, 1:].sum(axis=1).sort_values(ascending=False).head(10)
print("\nTop Manufacturers:\n", top_manufacturers)

```python
latest_year = 2024
market_share_2024 = df2.groupby("Maker")[str(latest_year)].sum()
market_share_2024 = (market_share_2024 / market_share_2024.sum()) * 100
print("\nMarket Share in 2024:\n" ,market_share_2024.sort_values(ascending=False).head(10))
```

### Total Sales Per Catogory

```python
category_performance = df2.groupby("Cat").sum().iloc[:, 1:].sum(axis=1)
print("\nTotal Salses Per Category:\n" ,category_performance)
```

### Biggest Yearly Fluctuation

```python
sales_variability = df2.iloc[:, 2:].std(axis=1).sort_values(ascending=False).head(10)
print("\nBiggest Sales Fluctuations:\n",sales_variability)
```

### Best And Worst Years For Sales

```python
best_year = total_sales_per_year.idxmax()
worst_year = total_sales_per_year.idxmin()
print("\nBest Year:\n",best_year,
"\nWorst Year:\n", worst_year)
```

### Recent Year Performance

```python
recent_performance = total_sales_per_year[["2022", "2023", "2024"]]
print("\nRecent Performance:\n",recent_performance)
```

# Key Performance indicators of the given datasets

### State With Maximum PCS

```python
state_max_pcs = df3.loc[df3["No. of Operational PCS"].idxmax()]
print("\nState with Max PCS:\n",state_max_pcs)
```

### State With Minimum PCS

```python
state_min_pcs = df3.loc[df3["No. of Operational PCS"].idxmin()]
print("\nState with Min PCS:\n", state_min_pcs)
```

### Top 5 States With Most PCS

```python
top_5_states = df3.nlargest(5, "No. of Operational PCS")
print("\nTop 5 States with Most PCS:\n", top_5_states)
```

### States With Least PCS

```python
least_5_states = df3.nsmallest(5, "No. of Operational PCS")
print("\n5 States with Least PCS:\n", least_5_states)
```

### Mean and Median PCS Per State

```python
mean_pcs = df3["No. of Operational PCS"].mean()
median_pcs = df3["No. of Operational PCS"].median()
print("\nMean PCS:\n",mean_pcs,
"\nMedian PCS:\n",median_pcs)
```

### Standard Deviation of PCS Distribution

```python
std_pcs = df3["No. of Operational PCS"].std()
print("Standard Deviation of PCS:",std_pcs)
```

### States Above Average PCS

```python
states_above_avg = df3[df3["No. of Operational PCS"] > mean_pcs]
print("\nStates Above Average PCS:\n", states_above_avg)
```
